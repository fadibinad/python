```python
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

# Load a dataset
iris = load_iris()
X, y = iris.data, iris.target

# Create a Random Forest Classifier
rf = RandomForestClassifier(n_estimators=100)  # You can adjust the number of trees

# Fit the Random Forest
rf.fit(X, y)

# Extract a tree from the forest (change the index as needed)
tree_to_visualize = 0

# Plot the decision tree
plt.figure(figsize=(20, 10))  # Adjust the figure size as needed
plot_tree(rf.estimators_[tree_to_visualize], filled=True, feature_names=iris.feature_names, class_names=iris.target_names, rounded=True)
plt.show()

```
