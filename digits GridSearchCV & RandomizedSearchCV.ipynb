```python
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_digits
digit=load_digits()
```

```python
dir(digit)
```

```python
inputs= digit.data
target= digit.target
```

```python
test={'svm':
      {'model':svm.SVC(gamma='auto'),
       'params':{'C':[1,10,40],'kernel':['rbf','linear']}
      },
      'rf':
      {'model':RandomForestClassifier(),
       'params':{'criterion':['gini','entropy'],'n_estimators':[10,20,40]}
      },
      'LogisticRegression': 
      {'model':LogisticRegression(),
       'params':{'solver':['liblinear','lbfgs'],'C':[1,10,20]}
      },
      'GaussianNB':
      {'model': GaussianNB(),
       'params':{}
      }, 
      'MultinomialNB':
      {'model':MultinomialNB(),
       'params':{}
      }, 
      'DecisionTreeClassifier':
      {'model': DecisionTreeClassifier(),
        'params': { 'criterion': ['gini','entropy']}
      }
     }  
```

```python
scores=[]
from sklearn.model_selection import RandomizedSearchCV
for mod, mp in test.items():
    clf= RandomizedSearchCV(mp['model'], mp['params'], cv=10, return_train_score=False)
    clf.fit(inputs, target)
    scores.append({'model': mod, 'best_score':clf.best_score_, 'best_parameter': clf.best_params_})
```

```python
import pandas as pd
res=pd.DataFrame(scores)
res
```
